```python
import numpy as np
import estdel
import estdel.constants as constants
```

```python
targets = np.arange(-constants.MAX_EST_MAG, constants.MAX_EST_MAG + constants.ESTIMATE_WIDTH, constants.ESTIMATE_WIDTH)
```

```python
v_ratios = np.exp(-2j * np.pi * (np.arange(constants.N_FREQS)*targets.reshape(targets.shape[0],1) + np.random.uniform(size =(targets.shape[0], 1))))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(np.angle(v_ratios), aspect = 'auto')
```

```python
predictor = estdel.VratioDelay(v_ratios)
```

```python
predictions = predictor.predict()
```

```python
np.mean(predictions - estdel.estdel._default_conversion_fn(targets) < 5)
```

```python
predictor = estdel.VratioDelaySign(v_ratios)
```

```python
predictions = predictor.predict()
```
